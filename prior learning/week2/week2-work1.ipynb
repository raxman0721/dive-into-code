# 【課題】統計基礎用語の説明

## 標準偏差

- 各データと平均の差の2乗したものが分散。
- 確率論的には確率変数から期待値を引いたものが分散
- 分散の平方根をとったものが標準偏差。
- ２つともデータの散らばり具合を表すが、**標準偏差の方が元データと同じ次元で比較出来る**ので便利なことが多い。
- **標本平均を使った分散は母集団よりも小さくなり、不偏推定量では無い！**
- 標準化する際に標準偏差で確率変数を割れば、分散が１になる。

## 共分散行列

- ベクトルの要素間の共分散の行列。
- 対角成分には分散、非対角成分は共分散が並ぶ
- 共分散が０のときは無相関となる
- 無相関であるからといって、統計的に独立（p(a,b) = p(a）p(b))であるとは言えない

## 正規分布について

- -∞　〜　∞の実数値を取る
- 平均値付近に確率密度が大きい。
- 平均値を中心として左右対称
- 中心極限定理によると、母集団分布にかかわらずサンプル数が増えると**標本平均の分布**は正規分布に近づいていく。
- １次元正規分布は平均μ、分散σ2
- 多次元正規分布は平均μ、分散はΣ（分散共分散行列）と表記される。
- 正規分布を仮定することで様々な事象の確率を求めることができる。検定や推定などにも使われる。

# 【課題】データ点の作成

```python
import numpy as np
import matplotlib.pyplot as pylab

#平均
mu1 = [-3,0]
mu2 = [0,-3]
#共分散
cov = [[1,0.8],[0.8,1]]

#500はデータ数
data1 = np.random.multivariate_normal(mu1,cov,500)
data2 = np.random.multivariate_normal(mu2,cov,500)
```

# 【課題】データ点へのラベル付け

```python
# それぞれのデータを行方向へ結合
data = np.concatenate([data1,data2])
data
```

```python
# 1と0の要素が入った列を作成。500番目までは１、それ以降は0
ones = np.ones(500)
zeros = np.zeros(500)

# dataと次元を合わせるためにreshapeして、結合
onezeros = np.concatenate([ones,zeros]).reshape((1000,1))
onezeros
```

```python
# dataとラベルを列方向へ結合
df = np.hstack((data,onezeros))
df
```

# 【課題】正しくラベル付けされているか見てみる

```python
# dfは[x,y,ラベル]の順で並んでるので、それをそのままプロット
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df[:,0], df[:,1], s=50, c=df[:,2], marker='o')
plt.show()
```
