# 【課題6】図の作成

- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する
- 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する

```python
import matplotlib.pyplot as plt
%matplotlib inline
# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
x = df.Species.value_counts()
plt.pie(x,autopct="%1.1f%%");
```

```python
import seaborn as sns

# 特徴量を一つ選び、boxplotを使ってラベルごとのデータの分布を可視化する
sns.boxplot(x=df.Species,y=df['petal length (cm)'])
```

```python
# - 特徴量を一つ選び、violinplotを使ってラベルごとのデータの分布を可視化する
sns.violinplot(x=df.Species,y=df['petal length (cm)']);
```

boxplotとviolinplotの違い
violinplotの方が情報量が多い。
violinplotの場合は全てのデータに対する確率密度が表示されている。
boxplotはいくつかの要約統計量しか表示されない為、violinplotの方がどの辺りにデータが密集しているか等が情報に含まれている。

# 【課題7】特徴量間の関係の確認

- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
plt.scatter(df['petal length (cm)'],df['petal width (cm)'],c=df.Species)
plt.xlabel('petal length')
plt.ylabel('petal width');
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
sns.pairplot(df,hue='Species');
```

```python
# ４つの特徴量に関する相関係数行列を作る
df_corr =  df.drop('Species',axis=1).corr()
df_corr
```

```python
# ヒートマップを作る
sns.heatmap(df_corr,annot=True)
```
